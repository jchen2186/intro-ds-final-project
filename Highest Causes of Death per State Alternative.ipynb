```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('raw_data/causes_death_state.csv')
```

```python
df.columns
```

```python
df.info()
```

```python
df = df.replace(to_replace='(S)', value=0.0).dropna()
```

```python
for column in df.columns[2:]:
    print(column)
    if df[column].dtype == object:
        df[column] = df[column].str.replace(',', '').dropna(axis=0).astype(np.float)
```

```python
# Finds the highest value column per row (Highest cause of death for each row aka state)
df['Highest Cause of Death'] = df.drop(['State', 'Total'], axis=1).idxmax(axis=1)
```
