```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("data/Influenza.csv")
df.info()
```

```python
df.describe()
```

```python
df['MMWR Year/Week'] = str(df['MMWR Year/Week'])
df['MMWR Year'] = df['MMWR Year/Week'].str[:4]
df['MMWR Week'] = df['MMWR Year/Week'].str[4:]
```

```python
sub_df = df[['State','Percent of deaths due to pneumonia or influenza']]

sub_df.sort_values(by='State')
sub_df.boxplot(by='State', figsize=(20,10),rot=90)
plt.ylabel('Percent of deaths due to pneumonia or influenza')
plt.show()
```

```python
ordered_states = list(sub_df.groupby('State').mean().reset_index().sort_values(by='Percent of deaths due to pneumonia or influenza')['State'])
print(ordered_states)
```
